```python
# Import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month 
x_axis = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis,y_axis)
```

```python
# create the ploy with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt() 
ax = plt.axes() 
ax.plot(x_axis, y_axis)
```

```python
# create the plot and add a label for the legend 
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label = 'Boston')

# create labels for the x and y axes 
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0-45 
plt.ylim(0, 45)

# set the x limit to jan - june 
plt.xlim('Jan', 'June')
# or plt.xlim(0, 5)

# create a title 
plt.title('PyBer Fare by Month')

# add a grid 
plt.grid()

# add the legend 
plt.legend()

# # change the figure size 
# plt.figure(figsize = (20, 10))
# or plt.subplots(figsize=(20, 10))
```

```python
# skill drill in 5.1.4 
# create object oriented chart 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, 'green', label='Boston', linewidth=2, marker= 'D')

# Change the marker to a diamond.

# Add a legend for the city of Boston.
ax.legend()

# Add a title and axes labels.
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Fare ($)')
ax.set_xlabel('Date')

# Add a y-axis limit.
ax.set_ylim(0,50)              

# Add grid lines.
ax.grid()
```

```python
# create the plot 
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Label the axis 
plt.xlabel('Data')
plt.ylabel('Fare($)')

# create title 
plt.title('PyBer Fare by Month')

# add legend 
plt.legend()
```

```python
# create plot 
plt.barh(x_axis,y_axis, color='magenta', label='Boston')

# invert the y axis data points 
plt.gca().invert_yaxis()

#labels & title 
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
```

```python
# create a vertical bar chart, object oriented method 
#create the play with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create horizontal bar chart, object-oriented

#creat the plot with ax.plt()
fig, ax = plt.subplots()

#to invert the chart just swap the variables inside the ax.barh below 
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

#labels 
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.legend()

# Invert the y axis 
ax.invert_yaxis()
```

```python
# two ways to create scatter plots 
# plt.plot(x_axis, y_axis, 'o')

plt.scatter(y_axis, x_axis, c='red', label='Chicago')
# set labels 
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare By Month')
#set x limit
plt.xlim(0,50)
# invert months 
plt.gca().invert_yaxis()
```

```python
#make the data in y axis bigger to make the dots bigger
# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data*3)

# Bubble charts - use format list compression to make dots bigger 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# scatter plot - object oriented 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create a bubble chart - object oriented 
fig, ax = plt.subplots()
# alpha means transparency 
ax.scatter(y_axis, x_axis, s = [i *5 for i in y_axis], c = 'skyblue', label='Boston', edgecolors='black', linewidth=2, alpha=.8)

# set lables 
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0, 50)
ax.legend()
# invert months 
ax.invert_yaxis()
```

```python
#Create pie chat 
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
#explode the pic chart and add % 
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# increase the size of the chart, this HAS to run before the plt.pie 
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# pie chart - object oriented 
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels = x_axis, shadow = True, 
       autopct='%.1f%%', explode=explode_values, startangle=100, counterclock = False,
      colors=colors)
plt.show()

#Reference for commands 
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html
    
## Skill Drill 
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.
```
